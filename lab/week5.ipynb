## Handling Missing Values

```python
#Replace the msiing values in Embarked(starting) column by 'S'
df['Embarked'].fillna('S', inplace=True)
#df['Sex'].fillna(df['Sex'].mode()[0], inplace=True)
```

```python
#Impute the missing values in Fare column with median value
df['Fare'].fillna(df['Fare'].median(),inplace=True)
```

```python
#Impute the Age column. It has too many missing values but it may be important. 
#We fill the missing values with numbers randomly drawn from distribution
#that has same mean and standard deviation as non missing Age data
#1. Count missing Age values
age_nan_count = df['Age'].isnull().sum()

# 2. Get the mean and standard deviation of the existing (non-missing) Age values
age_mean = df['Age'].mean()
age_std = df['Age'].std()

# 3. Generate random ages using the same mean and std
random_ages = np.random.normal(loc=age_mean, scale=age_std, size=age_nan_count)
df.loc[df['Age'].isnull(),'Age']=random_ages
```

## Feature Engineering - Creating new features

```python
# Create and add a nedfw feature FamilySize as a combination of SibSp and Parch
#FamilySize = Parch + SibSp + 1
df['FamilySize']=df['SibSp']+df['Parch']+1
```

```python
# Create new feature IsAlone from FamilySize
#If FamilySize == 1 then IsAlone = True else IsAlone = False
df['IsAlone']=df['FamilySize'] == 1
```

## Create a new attribute Has_cabin using the attribue Cabin. 
#if Cabin == NaN the Has_cabin = False else Has_cabin = True


```python
df['Has_cabin']=df['Cabin'].notna()
```

```python

#Drop the columns which may not be useful in prediction of survival
#Drop PassengerId, Name, Ticket, and Cabin
df.drop(['PassengerId','Name','Ticket','Cabin'],axis=1)
```

## Model Building

```python
#Prepare X and y
x = df.drop(columns='Survived')  # Features
y = df['Survived']  
print(x.shape)
print(y.shape)
```

```python
# Label encoding of categorical features
from sklearn.preprocessing import LabelEncoder

# Create separate LabelEncoders for each column
sex_encoder = LabelEncoder()
embarked_encoder = LabelEncoder()

# Apply encoding
df["Sex"] = sex_encoder.fit_transform(df["Sex"])
df["Embarked"] = embarked_encoder.fit_transform(df["Embarked"])
df
```

```python
# concat df_categorical with original df
categorical=pd.get_dummies(df[['Sex','Embarked']],drop_first=True)
df=pd.concat([df,categorical],axis=1)
df
```

```python
#Do a pairplot of all features

# sns.pairplot(df, hue='Survived', palette='pastel')
# plt.suptitle("Pairplot of All Features", y=1.02)
# plt.show()
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(20,20))
sns.pairplot(df,hue="Survived")
plt.show()
```

```python
# Drop non-numeric columns
df = df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin', 'Sex'], axis=1, errors='ignore')
```

```python
# Handle missing values instead of dropping everything
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# Encode categorical features
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df['Embarked'] = df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})

# Drop unused text columns
df = df.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'], errors='ignore')

# Features and target
x = df.drop(columns='Survived')
y = df['Survived']

# Check shape before splitting
print("x shape:", x.shape)
print("y shape:", y.shape)

# Split only if there are rows
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=42)
```

```python
x = df.drop(columns='Survived')  # Features
y = df['Survived'] 
```

```python
# Splitting the data into train and validation set - 90% train 10% validation
#X_train, X_test, y_train, y_validation
from sklearn.model_selection import train_test_split

# Assuming you have X (features) and y (target/label)
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.1, random_state=42
)

# Optional: print shapes to confirm split

print("X_train shape:", x_train.shape)
print("X_val shape:", x_test.shape)
print("y_train shape:", y_train.shape)
print("y_val shape:", y_test.shape)
```

```python
# Create the model
# Fitting the decision tree with default hyperparameters
# Importing decision tree classifier from sklearn library
# Import necessary 

from sklearn.tree import DecisionTreeClassifier

# Create the model with default parameters
model = DecisionTreeClassifier(criterion='entropy')

# Fit the model on training data
model.fit(x_train, y_train)
```

```python
#Print the tree

# Importing required packages for visualization
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
plt.figure(figsize=(20,20))
plt.figure(figsize=(20,10))
plot_tree(model,filled=True,feature_names=x.columns, class_names=["Not Survived","Survived"])
plt.show()
```

```python
#Print the tree
# Importing required packages for visualization
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
plt.figure(figsize=(10,12))
plot_tree(model,filled=True,class_names=['Survived','NotSurvived'],feature_names=df.columns,precision=2,fontsize=13,max_depth=2)
plt.show()
```
